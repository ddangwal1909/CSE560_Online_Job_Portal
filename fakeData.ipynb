```python
company_db = []
state_country_list=[]
companies=[]
emails=[]
with open('companies_sorted.csv','r') as fl:
    cnt=0
    for lines in fl:
        #print(lines)
        if cnt==0:
            cnt+=1
        elif cnt==100:
            break
        else:
            try:
                cid=cnt
                #print(lines.split('"')[2].split(','))
                cname=lines.split(',')[1]
                cemail='admin@'+lines.split(',')[2]
                cstate=lines.split('"')[1].split(',')[-2]
                ccountry = lines.split('"')[2].split(',')[1]
                company_db.append((cid,cname,cemail,cstate,ccountry))
                companies.append(cname)
                emails.append((cemail,1))
                state_country_list.append((cstate,ccountry))
                cnt+=1
            except:
                cnt=cnt

f = open('companydata.csv','a')
f.write("cid,cname,cemail,cstate,ccountry\n")
for i in company_db:
    for j in range(4):
        f.write(str(i[j]).strip()+',')
    f.write(i[-1]+'\n')
f.close()
    
```

```python
import requests
from random import sample
import random
import time

# Prints list of random items of given length
list1 = ['HighSchool','Undergraduate','Masters','PhD']   
listg = ['Male','Female','Other']
listrace = ['Asian','Hispanic Latino','African American','White','Black','Other']
listWork = ['Yes','No']
listsalary = ['<=50k','>50k - <=100k','>100k - <=150k','>150k - <=200k','>200k']
applicantdb = []
cnt=1
for i in range(1000):
    print(i)
    time.sleep(0.5)
    r = requests.get('https://api.namefake.com/')
    data = r.json()
    appid = cnt
    fname = data['name'].split()[0]
    lname = data['name'].split()[1]
    email = data['email_u']+'@'+data['email_d']
    education = str(sample(list1,1)[0])
    age = random.randint(21, 50)
    race = sample(listg,1)[0]
    work_authorization = sample(listWork,1)[0]
    tmp=sample(state_country_list,1)[0]
    state=tmp[0]
    country=tmp[1]
    expected_salary = sample(listsalary,1)[0]
    applicantdb.append((appid,fname,lname,email,education,age,race,work_authorization,state,country,expected_salary))      
    cnt+=1

f = open('applicant_data.csv','a')
f.write("appid,fname,lname,email,education,age,race,work_authorization,state,country,expected_salary\n")
for i in applicantdb:
    for j in range(10):
        f.write(str(i[j]).strip()+',')
    f.write(i[-1]+'\n')
f.close()

```

```python
######## job fake data generation ######

lst_jobs = ['Human Resource Manager','Web Developer','Database Administrator','Engineering Manager', 'Analyst-Product Development', 'Frontend Developer','Hardware Engineer','SDE-1','SDE-2','SDE-3','Algorithmic Coder']

job_db=[]

for i in range(10000):
    jobid=i+1
    jobtitle=sample(lst_jobs,1)[0]
    company=sample(companies,1)[0]
    salary=sample(listsalary,1)[0]
    educationexperience=sample(list1,1)[0]
    job_db.append((jobid,jobtitle,company,salary,educationexperience))


f = open('job_listing.csv','a')
f.write("jobid,jobtitle,company,salary,education\n")
for i in job_db:
    for j in range(3):
        f.write(str(i[j]).strip()+',')
    f.write(i[-1]+'\n')
f.close()
```

```python
##### PASSWORD GENERATION 
password_db=[]
import string
def get_random_string(length):
    # choose from all lowercase letter
    letters = string.ascii_lowercase
    result_str = ''.join(random.choice(letters) for i in range(length))
    #print("Random string of length", length, "is:", result_str)
    return result_str

for i in emails:
    emailid = i[0]
    companyflag=i[1]
    password=get_random_string(8)
    password_db.append((emailid,companyflag,password))

f = open('passwords_data.csv','a')
f.write("emailid,companyflag,password\n")
for i in password_db:
    for j in range(2):
        f.write(str(i[j]).strip()+',')
    f.write(i[-1]+'\n')
f.close()
```
